# Feature Elmination

#Some of feature showing complete corrlation with other feature so we will eliminate  those: such as area mean, premeter mean,radius worst, texture worst, concavity mean, , perimatere worst, texture worst,area se,concavity mean...


```python
df_new=df.drop(["area_mean",'perimeter_mean',"radius_worst",'area_worst',
         'perimeter_worst','texture_worst','concavity_mean','perimeter_se', 'area_se'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encod_y=LabelEncoder()
df_new.loc[:,"diagnosis"]=label_encod_y.fit_transform(df_new.loc[:,"diagnosis"].values)

df_new.head()
```

```python
x=df_new.drop("diagnosis",axis=1)
y=df_new.iloc[:,0]
y
```

# Building Model

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=8)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)
x_test_std
```

# Principal Component Analysis

```python
from sklearn.decomposition import PCA

pca=PCA().fit(x_train_std)


#Finiding Cumulative sum of Varaiance: 
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("N_components")
plt.ylabel("Explained Variance")
# First 10 components explained above 92% of Variance in Data...
```

```python
#Selecting first 10 components...
pca=PCA(n_components=10)

x_train_pca=pca.fit_transform(x_train_std)
x_test_pca=pca.transform(x_test_std)


print(x_train_pca.shape)
print(y_train.shape)
```

# Applying KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier

k=range(1,40)

mean_accuracy_training=[]
mean_accuracy_testing=[]


for i in k:
    classifier=KNeighborsClassifier(n_neighbors=i)
    classifier.fit(x_train_pca,y_train)
    mean_accuracy_training.append(classifier.score(x_train_pca,y_train))
    mean_accuracy_testing.append(classifier.score(x_test_pca,y_test))
    

    
    
sns.lineplot(k, mean_accuracy_training,label="Training Score",legend='brief')

sns.lineplot(k,mean_accuracy_testing,label="Tesing Score",legend='brief')


plt.ylabel("Mean Accuracy (%)")
plt.xlabel("n_neighbors Values")

plt.show()

```

```python

from sklearn.neighbors import KNeighborsClassifier


error=[]


for c in range(1,20):
    classifier=KNeighborsClassifier(n_neighbors=c)
    classifier.fit(x_train_pca,y_train)
    predict=classifier.predict(x_test_pca)
    error.append(np.mean(predict!=y_test))

plt.figure(figsize=(12, 6))
plt.plot(range(1, 20), error, color='green', linestyle='dashed', marker='o',markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel("Mean Error")
```

```python
#Best Value Of K found is 6:
from sklearn.neighbors import KNeighborsClassifier


classifier=KNeighborsClassifier(n_neighbors=6)
classifier.fit(x_train_pca,y_train)
predict=classifier.predict(x_test_pca)
#print('Misclassification rate: ',(y_test_pca != predict).sum())
print("Testing Accuracy of Model:    ", classifier.score(x_test_pca,y_test))
print("Training Accuracy of Model:    ", classifier.score(x_train_pca,y_train))

```

```python
x_test_pca.shape
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm=confusion_matrix(y_test,predict)
cr=classification_report(y_test,predict)

print(sns.heatmap(cm,annot=True,fmt=".2f",cmap="coolwarm"))
print(cr)
```

# Decesion Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


d_clf=DecisionTreeClassifier(random_state=8)

d_clf.fit(x_train_pca,y_train)


from sklearn.model_selection import cross_val_score


score_dtrain = np.mean(cross_val_score(d_clf,  x_train_pca, y_train, scoring='accuracy'))
score_dtest = np.mean(cross_val_score(d_clf, x_test_pca, predict, scoring='accuracy'))

print("Accuracy for Decision Tree on Training(CV):",(score_dtrain))
print("Accuracy for Decision Tree on Testing(CV):",score_dtest)
```

```python
predict_y_dec=d_clf.predict(x_test_pca)

cm_d=confusion_matrix(y_test,predict_y_dec)
cr_d=classification_report(y_test,predict_y_dec)
print(cr_d)
print(sns.heatmap(cm_d,annot=True,cmap="coolwarm"))
```

# Support Vector Machines

```python
from sklearn.svm import SVC

svc=SVC(random_state=8,probability=True)
scores_svc = np.mean(cross_val_score(svc, x_test_pca, y_test, scoring='accuracy'))
p_score_svc = np.mean(cross_val_score(svc,x_test_pca,y_test, scoring='precision'))
r_score_svc = np.mean(cross_val_score(svc, x_test_pca, y_test, scoring='recall'))
print("Accuracy for svc:",scores_svc)
print ("Precision for svc", p_score_svc)
print ('Recall score for svc',r_score_svc)




svc.fit(x_train_pca,y_train)
predict_y_svc=svc.predict(x_test_pca)
cm_d=confusion_matrix(y_test,predict_y_svc)
cr_d=classification_report(y_test,predict_y_svc)
print(cr_d)
print(sns.heatmap(cm_d,annot=True,cmap="coolwarm"
                 ,fmt=".0f",xticklabels = ['Benign', 'Malignant'], yticklabels = ['Benign', 'Malignant']))


plt.ylabel('True')
plt.xlabel('Predicted')
plt.title("Confusion Matrix for SVM")
```

# KMEANS MODEL

```python
from sklearn.cluster import KMeans

wscc=[]

for k in range(1,15):
    kmeans=KMeans(n_clusters = k)
    kmeans.fit(x)
    wscc.append(kmeans.inertia_)
plt.plot(range(1,15),wscc)
plt.xlabel("K Values")
plt.ylabel("WSCC")
```

```python
df_nolabels=x.copy()
#plotting Data After K=2
kmean=KMeans(n_clusters=2)
cluster=kmean.fit_predict(df_nolabels)
#df_nolabels["cluster"]=cluster
df_nolabels.head()

df_nolabels["cluster"] = cluster
df_nolabels["cluster"].unique()
```

```python
plt.figure(figsize=(10,10))


for_zero=df_nolabels[df_nolabels["cluster"]==0]
for_one=df_nolabels[df_nolabels["cluster"]==1]

plt.scatter(for_zero["radius_mean"],for_zero["texture_mean"],color='red')
plt.scatter(for_one["radius_mean"],for_one["texture_mean"],color="green")
plt.xlabel('radius_mean')
plt.ylabel("texture_mean")
```

```python
# Data centroids middle of clustered scatters

plt.figure(figsize=(15,10))


plt.scatter(df_nolabels["radius_mean"],df_nolabels["texture_mean"],c=cluster)
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],color="red",alpha=1)
plt.xlabel('radius_mean')
plt.ylabel("texture_mean")
```
